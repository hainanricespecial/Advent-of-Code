```python
# Part 1
# Initialize the dial state.
dial_state = 50

# Read the file.
inputfile = open("DAY1INPUT.txt")

bingo = 0

for line in inputfile:
    direction = line[0]
    steptakenextract = int(line[1:])

    # Do the calculation
    if(direction == "R"):
        dial_state = (dial_state + steptakenextract) % 100
    
    if(direction == "L"):
        dial_state = (dial_state - steptakenextract) % 100

    if(dial_state == 0):
        bingo = bingo + 1


print(bingo)
    
```

```python
# The Heart, Part 2

# Initialize the dial state.
dial_state = 50

# Read the file.
inputfile = open("DAY1INPUT.txt")

bingo = 0

for line in inputfile:
    direction = line[0]
    steptakenextract = int(line[1:])

    # Don't tell lies about me, I won't tell truths about you
    if(steptakenextract > 100):
        remainder = steptakenextract % 100
        extrarotation = (steptakenextract - remainder) / 100
        bingo = bingo + extrarotation
        steptakenextract = remainder

    # Do the calculation
    if(direction == "R"):
        dial_state = dial_state + steptakenextract

        if(dial_state > 100):
            dial_state = dial_state % 100
            bingo = bingo + 1
        
    
    if(direction == "L"):
        dial_state = dial_state - steptakenextract

        if(dial_state < 0):
            dial_state = dial_state % 100
            bingo = bingo + 1

    if(dial_state == 0):
        bingo = bingo + 1


print(bingo)
```
